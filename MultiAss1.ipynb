```python
Q4.write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes.
Here's a Python program that creates two threads. Thread one prints a list of squares, and thread two prints a list of cubes.
```

```python
import threading

def print_squares():
    for i in range(1, 11):
        print(i ** 2)

def print_cubes():
    for i in range(1, 11):
        print(i ** 3)

t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)

t1.start()
t2.start()

t1.join()
t2.join()

print("Done")
```

```python
In this program, we define two functions: print_squares and print_cubes. Each function uses a for loop to print a list of squares or cubes, 
respectively.

We then create two Thread objects, t1 and t2, and assign each a target function using the target parameter. We then start both threads using
the start method.

Finally, we use the join method to wait for both threads to finish before printing "Done". This ensures that the program does not exit before 
both threads have completed their tasks.
```

```python
Q5. State advantages and disadvantages of multithreading
ans:
Multithreading is a programming technique that involves running multiple threads of execution within a single program. Here are some advantages and 
disadvantages of multithreading:

Advantages:

1. Increased performance: Multithreading can improve the performance of certain types of programs by allowing multiple tasks to run concurrently. 
This can be especially useful for programs that perform I/O-bound operations, such as file or network access, as other threads can continue to
execute while waiting for I/O operations to complete.

2. Improved responsiveness: By separating a program's tasks into multiple threads, the user interface can remain responsive even if one thread is
busy performing a time-consuming task.

3. Better resource utilization: Multithreading can help maximize the use of available system resources by allowing multiple tasks to execute
  simultaneously.

4. Simplified code: In some cases, multithreading can simplify code by breaking a complex problem into smaller, more manageable parts that 
can be executed independently.

Disadvantages:

1. Complexity: Multithreading can add complexity to a program, making it more difficult to understand and debug.

2. Synchronization issues: When multiple threads access shared resources, synchronization issues can arise. These issues can lead to bugs
   that are difficult to detect and fix.

3. Overhead: Creating and managing threads can involve overhead, such as additional memory and CPU usage, which can impact performance.

4.Debugging difficulties: Debugging multithreaded code can be more challenging than debugging single-threaded code, as issues can occur in 
 non-deterministic ways.

Overall, multithreading can be a powerful tool for improving performance and responsiveness in certain types of programs, but it also requires 
careful consideration of the trade-offs involved and careful attention to synchronization and debugging issues.
```

```python
Q6. Explain deadlocks and race conditions.
Ans:
Deadlocks and race conditions are two common issues that can occur when using multithreading.

Deadlocks: A deadlock is a situation where two or more threads are blocked waiting for each other to release resources that they need in 
order to proceed. For example, Thread 1 may hold a lock on Resource A and be waiting for Resource B, while Thread 2 may hold a lock on Resource
B and be waiting for Resource A. Neither thread can proceed until the other releases its lock, resulting in a deadlock. Deadlocks can be difficult 
to detect and resolve, and can cause a program to become unresponsive.

Race conditions: A race condition occurs when two or more threads access a shared resource in a way that leads to unexpected behavior. For example, 
if two threads are trying to increment the same variable at the same time, the value of the variable may not be what is expected due to the 
non-deterministic order in which the threads execute. Race conditions can lead to bugs that are difficult to detect and reproduce, and can cause a 
program to behave in unexpected ways.

To prevent deadlocks, it is important to avoid circular waits, which occur when a thread is waiting for a resource that is held by another
thread that is, in turn, waiting for a resource that is held by the first thread. To prevent race conditions, it is important to use synchronization 
mechanisms, such as locks or semaphores, to ensure that only one thread can access a shared resource at a time. Synchronization mechanisms can also
be used to ensure that threads access shared resources in a specific order, which can help prevent race conditions.
```
